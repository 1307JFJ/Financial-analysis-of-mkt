```python
import numpy as np
import pandas as pd

import statsmodels.api as sm
import statsmodels.stats.api as sms
import statsmodels.discrete.discrete_model as smdiscrete
```

```python
pd.set_option('use_inf_as_na', True) 
# 需要将inf替换成nan，或者在文件开头加入设置语句
```

```python
stock_meta = pd.read_csv('stock_metadata.csv')
stock_meta
```

```python
fund = pd.read_csv('corp_fund.csv')

fund = fund[fund['dimension'] == 'ARQ']

fund['datekey'] = pd.to_datetime(fund['datekey'])
```

```python
df_left = pd.merge(fund, stock_meta, on='ticker', how='left')

df_left = df_left.set_index('datekey')

df_left.head()
```

## 1. Fixed effects example code

- Let's introduce an indicator variable (“dummy variable”) for each industry.

```python
industry_dummies = pd.get_dummies(df_left['sicsector'])

industry_dummies.sum() #purely for exploring the data, has no other purpose

industry_dummies.describe() #purely for exploring the data, has no other purpose

industry_dummies
```

```python
# Next merge the dummy variable dataframe into the main dataframe.

data_w_dummies = pd.concat([df_left, industry_dummies],
                           axis=1)
```

```python
# Then drop 1 dummy variable, which will be the control group

data_w_dummies.drop(['Wholesale Trade'],
                    inplace=True,
                    axis=1)
```

```python
# Generate our dependent variable(Y) (E/P ratio)
# independent variable(X)operating margin

data_w_dummies['EP_ratio'] = data_w_dummies['eps'] / data_w_dummies['price']

data_w_dummies['Operating_margin'] = data_w_dummies['opinc'] / data_w_dummies['revenue']

data_w_dummies
```

> **Let's compare OLS regression with and without dummy variables.**

```python
# OLS regression without dummy variables

X1 = sm.add_constant(data_w_dummies[['Operating_margin']])

y = data_w_dummies['EP_ratio']

model1 = sm.OLS(y, X1, missing='drop')

result1 = model1.fit()

result1.summary()

# get result with no fixed effects
```

```python
# OLS regression with dummy variables

X2 = sm.add_constant(data_w_dummies[['Operating_margin',
                                     'Agriculture Forestry And Fishing',
                                     'Construction',
                                     'Finance Insurance And Real Estate',
                                     'Manufacturing', 
                                     'Mining','Retail Trade',
                                     'Services',
                                     'Transportation Communications Electric Gas And Sanitary Service']])
```

```python
model2 = sm.OLS(y, X2, missing='drop').fit()

result2 = model2.summary()

result2
# get result with fixed effects
```

- We note t-statistic for estimated coefficient on‘operating_margin’ is very low.

> Next  Let's generate log versions of both dependent and independent variables

- Run regression with log variables


```python
data_w_dummies['log_operating_margin'] = np.log(data_w_dummies['Operating_margin'])

X3 = sm.add_constant(data_w_dummies['log_operating_margin'])

# only included X is the log style
model3 = sm.OLS(y, X3, missing='drop').fit()

result3 = model3.summary()

result3
```

```python
# let's change Y into the log style
data_w_dummies['log_EP_ratio'] = np.log(data_w_dummies['EP_ratio'])

log_y = data_w_dummies['log_EP_ratio']

model4 = sm.OLS(log_y, X3, missing='drop').fit()

result4 = model4.summary()

result4   # log variables, no Fixed Effect
```

- Same regression, but with dummy variables included

```python
X4 = sm.add_constant(data_w_dummies[['log_operating_margin',
                                     'Agriculture Forestry And Fishing',
                                     'Construction',
                                     'Finance Insurance And Real Estate',
                                     'Manufacturing',
                                     'Mining',
                                     'Retail Trade',
                                     'Services',
                                     'Transportation Communications Electric Gas And Sanitary Service']])

model5 = sm.OLS(log_y, X4, missing='drop').fit()

result5 = model5.summary()

result5 # log variables, with Fixed Effect
```

- **We note t-statistic is significantly improved, but could this be a result of intra-cluster correlations?**

    -  Observations within same industry are likely to be highly correlated versus observations in different industries, which violates OLS assumption of iid datapoints.
    
    -  Estimates may therefore greatly overstate efficiency (inflated t-statistics, artificially low pvalues, downward biases standard errors)
    
> Let's adjust the standard error estimates by clustering errors within each industry

```python
# clustering standard variables
data_w_dummies.dropna(subset=['log_EP_ratio','log_operating_margin'],
                      inplace=True)

#because of a bug in python where fillna is not working perfectly
#note that we cannot cluster by str variables in python, hence using siccode instead of sicsector
```

```python
# 对model5的相关参数进行调整，得到model6
model6 = sm.OLS(log_y, X4, missing='drop').fit(cov_type='cluster',
                                               cov_kwds={'groups': data_w_dummies['siccode']})

result6 = model6.summary()

result6
```

> **Estimated t-statistic is several times lower after clustering!**

```python
# Using a binary dependent variable
# generate categorical variable for probit/logit analysis

data_w_dummies['pay_dividend'] = data_w_dummies['dps']>0
data_w_dummies['pay_dividend']
data_w_dummies['pay_dividend'].mean()
#formatting the data for estimation models
data_w_dummies['pay_dividend'] = data_w_dummies['pay_dividend'].astype(int)

# try using OLS anyway
# treat "pay_dividend" as y
model7 = sm.OLS(data_w_dummies['pay_dividend'], 
                sm.add_constant(data_w_dummies[['log_operating_margin',
                                                'Agriculture Forestry And Fishing',
                                                'Construction',
                                                'Finance Insurance And Real Estate',
                                                'Manufacturing',
                                                'Mining','Retail Trade',
                                                'Services','Transportation Communications Electric Gas And Sanitary Service']]),
                missing='drop').fit(cov_type='cluster',
                                    cov_kwds={'groups':data_w_dummies['siccode']})

result7 = model7.summary()
result7
```

```python
# use logit instead
logit_model = smdiscrete.Logit(data_w_dummies['pay_dividend'],
                               sm.add_constant(data_w_dummies[['log_operating_margin',
                                                               'Agriculture Forestry And Fishing',
                                                               'Construction',
                                                               'Finance Insurance And Real Estate',
                                                               'Manufacturing',
                                                               'Mining','Retail Trade',
                                                               'Services','Transportation Communications Electric Gas And Sanitary Service']]),
                               missing='drop').fit(cov_type='cluster',
                                                   cov_kwds={'groups':data_w_dummies['siccode']})

logit_result = logit_model.summary()

logit_result
```

## PCA

```python
from sklearn.decomposition import PCA

data = data_w_dummies

numerator = ['cashneq', 'debt', 'ebit', 
             'ebt', 'eps', 'equity', 'fcf', 
             'gp', 'inventory', 'liabilities', 
             'payables', 'receivables', 'tangibles', 'workingcapital']

denominator = ['assets','revenue']

features_list = []
```

```python
for n in numerator:
    for d in denominator:
        tag = n + '_' + d
        data[tag] = np.log(data[n] / data[d])
        features_list.append(tag)
```

```python
data.dropna(subset=features_list,
            inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
features = data.loc[:, features_list].values

features = StandardScaler().fit_transform(features)
```

```python
pca = PCA(n_components=4)

princinpal_components = pca.fit_transform(features)

princinpal_components

pca.explained_variance_ratio_
```

```python
pc_df = pd.DataFrame(princinpal_components)

pc_df.columns = ['PC1','PC2','PC3','PC4']
pc_df.index = data.index

pc_df.corr()
```

```python
data_merge = pd.concat([data, pc_df],
                       axis=1)

n_d_model = sm.OLS(data_merge['log_EP_ratio'],
                   sm.add_constant(data_merge[features_list]),
                   missing='drop').fit(cov_type='cluster',
                                       cov_kwds={'groups':data_w_dummies['siccode']})

n_d_result = n_d_model.summary()

n_d_result
```

```python
pca_model = sm.OLS(data_merge['log_EP_ratio'],
                   sm.add_constant(data_merge[['PC1','PC2','PC3','PC4']]),
                   missing='drop').fit(cov_type='cluster',
                                       cov_kwds={'groups': data_w_dummies['siccode']})

pca_result = pca_model.summary()

pca_result
```

```python
#lets figure out what 'PC1' means
features_list.append('PC1')
data_merge[features_list].corr()
```

- Looking just at raw variables with >40% correlation to PC1, we can hypothesize that it relates to “working capital / cash conversion cycle management” of the firm.


### INTERACTION VARIABLES

```python
# Generate the interaction variables. I
# we create interaction variables as log_operating_margin*industry dummy
dummies = industry_dummies.columns.values

dummies = dummies[1 : len(dummies) - 1]

dlist = []

for i in dummies:
    tag = 'log_operating_margin' + '_' + i
    data[tag] = data['log_operating_margin'] * (data[i])
    dlist.append(tag)
    dlist.append(i)
    
dlist.append('log_operating_margin')
```

```python
# Run OLS estimation with interaction variables
result = sm.OLS(data['log_EP_ratio'],
                sm.add_constant(data[dlist]),
                missing='drop').fit(cov_type='cluster',
                                    cov_kwds={'groups': data_w_dummies['siccode']})

result.summary()
```

## SQUARED VARIABLES

- We hypothesize that a stock's valuation has a concave relationship to its asset tangiblility.

```python
# Generate measures of asset tangibility and squared asset tangibility
data['tangibles_assets_normal'] = data['tangibles'] / data['assets']

data['tangibles_assets_sq'] = data['tangibles_assets_normal'] * data['tangibles_assets_normal']
```

```python
# Run OLS versus against asset tangibility and asset tangibility_sq
result = sm.OLS(data['EP_ratio'],
                sm.add_constant(data[['tangibles_assets_normal',
                                      'tangibles_assets_sq']]),
                missing='drop').fit(cov_type='cluster',
                                    cov_kwds={'groups':data_w_dummies['siccode']})

result.summary()
```

- Non-linear relationship has been seen through a squared term！

```python
# Run OLS versus only asset tangibility
result = sm.OLS(data['EP_ratio'],
                sm.add_constant(data[['tangibles_assets_normal']]),
                missing='drop').fit(cov_type='cluster',
                                    cov_kwds={'groups':data_w_dummies['siccode']})

result.summary()
```

- A univariate regression masks a more non-linear relationship.

## LAG INDEPENDENT VARIABLES

```python
# we set the index of the observations to the tuple (ticker, date).
# This is because when we compute lags, we only want to lag observations‘within’ each ticker
data_merge.index = [data_merge['ticker'],
                    data_merge.index]

# lags the observations by 1 quarter
data_lag1 = data_merge.groupby(level=0).shift(1)

dataset = pd.concat([data_merge['log_EP_ratio'],
                     data_lag1[['PC1','PC2','PC3','PC4']]],
                     axis=1)

dataset.dropna(inplace=True)
```

```python
result = sm.OLS(dataset['log_EP_ratio'],
                sm.add_constant(dataset[['PC1','PC2','PC3','PC4']]),
                missing='drop').fit()

print(result.summary())
```

- We note that the coefficient on PC1 has changed sign!

## F-TESTS

```python
from statsmodels.formula.api import ols

formula = 'log_EP_ratio ~ tangibles_assets_normal + tangibles_assets_sq + cashneq_assets + gp_assets + gp_revenue + fcf_assets'

result = ols(formula, data).fit()

hypothesis_1 = '(tangibles_assets_normal = cashneq_assets), tangibles_assets_sq = -0.5'

f_test = result.f_test(hypothesis_1)

print(f_test)
```

```python
hypothesis_2 = '(gp_assets = gp_revenue)'

f_test2 = result.f_test(hypothesis_2)

print(f_test2)
```

## Hazard models

```python
survival = pd.read_csv('survival_data.csv')


#drop data where delisted status is unsure
survival.dropna(subset=['isdelisted'],
                inplace=True)


# code a 1/0 variable based on whether delisted or not.
# Python needs a numerical value
survival.loc[survival['isdelisted'] == 'N', 'dead'] = 0
survival.loc[survival['isdelisted'] == 'Y', 'dead'] = 1

# estimate probabiltiy of delisting based on revenue category
size_dummies = pd.get_dummies(survival['scalerevenue'])

survival_df = pd.concat([survival, size_dummies],
                         axis=1)
```

```python
import statsmodels.formula.api as smf

survival_df.rename(columns={'1 - Nano':'Nano',
                            '2 - Micro': 'Micro',
                            '3 - Small': 'Small',
                            '4 - Mid': 'Mid',
                            '5 - Large': 'Large',
                            '6 - Mega': 'Mega'},
                   inplace=True)
```

```python
mod = smf.phreg('duration ~ 0 + Nano + Micro + Small + Mid + Large + Mega',
                survival_df,
                status = survival_df['dead'].values,
                ties='efron')

result = mod.fit()

result.summary()
```

```python
survival_df['sizecat'] = 0
survival_df.loc[survival_df['Micro']==1, 'sizecat'] = 1
survival_df.loc[survival_df['Small']==1, 'sizecat'] = 2
survival_df.loc[survival_df['Mid']==1, 'sizecat'] = 3
survival_df.loc[survival_df['Large']==1, 'sizecat'] = 4
survival_df.loc[survival_df['Mega']==1, 'sizecat'] = 5
survival_df.sizecat
mod = smf.phreg("duration ~ 0 + sizecat", survival_df, status = survival_df["dead"].values, ties = "efron")
rslt = mod.fit()
rslt.summary()
```
